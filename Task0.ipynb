## Remove all the NULL values from the dataset.

## Remove 'Name' and 'PassengerID' column

```python
test_dset2=test_dset1.drop(labels={'Name','PassengerId'},axis=1)
test_dset2.head()
```

```python
train_dset2=train_dset1.drop(labels={'Name','PassengerId'},axis=1)
train_dset2
```

## Plot a histogram for 'Fare','Age' column
